# Ajuste para minimos cuadrados usando las funciones de numpy para realizar el least square method 'lstsq()'

```python
import numpy as np
```

```python
a=np.load('fit_exercise.npy')
a
```

a[0] son las X, a[1] son las Y, y a[2] son los errores en Y.

```python
A = np.vstack([a[0], np.ones(len(a[0]))]).T
m, c = np.linalg.lstsq(A,a[1],rcond=None)[0]
m,c
```

```python
import matplotlib.pyplot as plt
plt.plot(a[0], a[1], 'o', label='Original data', markersize=3)
plt.errorbar(a[0], a[1],yerr= a[2],ls='',label='both limits (default)')
plt.plot(a[0], m*a[0] + c, 'r', label='Fitted line')
plt.legend()
plt.show()
```

# Ahora sin usar librerias para minimos cuadrados:

```python
#obtenemos el numero total de datos n
n=len(a[0])
print(n)
#obtenemos la media de los X
x_mean=np.mean(a[0])
#obtenemos la suma total de cada Xi elevado al cuadrado
sq= a[0]*a[0]
x2_sum=sq.sum()
#obtenemos la suma de multimpliacar cada elemento Xi con cada Yi
sq2=a[0]*a[1]
xy_sum=sq2.sum()
#la media de los valores de Y
y_mean=np.mean(a[1])

c2= ((y_mean* x2_sum)- (x_mean * xy_sum) )/(x2_sum - (n*(x_mean)**2))
print(c2,c)

m2= (xy_sum -n*y_mean*x_mean)/(x2_sum - n*(x_mean**2))
print(m,m2)
```

```python
import matplotlib.pyplot as plt
plt.plot(a[0], a[1], 'o', label='Original data', markersize=3)
plt.errorbar(a[0], a[1],yerr= a[2],ls='',label='both limits (default)')
plt.plot(a[0], m*a[0] + c, 'b', label='Fitted line')
plt.plot(a[0], m2*a[0] + c2, 'r', label='Fitted line2')
plt.legend()
plt.show()
```
